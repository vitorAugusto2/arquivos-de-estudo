### 2. Gr√°ficos

- **Funcoes gr√°ficas**: Cada gr√°fico que pode ser feito no matplotlib possui uma fun√ß√£o dedicada. Para alterar o tipo de gr√°fico, alteramos a fun√ß√£o gr√°fica de onde estamos chamando `plt`

| Tipo de Gr√°fico      | C√≥digo do gr√°fico  |
|----------------------|--------------------|
| Gr√°fico de linha     | plt.plot()         |
| Gr√°fico de dispers√£o | plt.scatter()      |
| Gr√°fico de barras    | plt.bar()          |
| Gr√°fico de pizza     | plt.pie()          |
| Histograma           | plt.hist()         | 

- **gr√°fico de linhas**: mostra mudan√ßas cont√≠nuas, geralmente usado para medir mudan√ßas ao longo do tempo
- **gr√°fico de dispers√£o**: usa posi√ß√£o para mostrar a rela√ß√£o, ou correla√ß√£o, entre dois valores num√©ricos
- **gr√°fico de barras**: usa a altura da barra para comparar uma medida entre vari√°veis ‚Äã‚Äãcateg√≥ricas
- **gr√°fico de pizza**: mostra-nos a divis√£o de um todo em suas partes
- **histograma**: mostra como um tipo de dados √© distribu√≠do

### 3. Anatomia do c√≥digo matplotlib

- **Funcoes gerais**: alterar aspectos gerais do gr√°fico, como t√≠tulo, legenda ou escala de eixo e r√≥tulos. Chamamos essas fun√ß√µes gerais porque elas se aplicam a todos os tipos de gr√°ficos

| Adicione um titulo         | plt.title()                 |   
|----------------------------|-----------------------------|
| Adicione uma legenda       | plt.legend()                |
| Adicionar r√≥tulos de eixo  | plt.ylabel(), plt.xlabel()  |
| Ajustar eixos              | plt.xscale(), plt.yticks()  |
| Imprimir gr√°fico           | plt.show()                  |

### 1. Gr√°fico de barra

Os gr√°ficos de barras usam o comprimento das barras para comparar dados categ√≥ricos.

A fun√ß√£o do gr√°fico do gr√°fico de barras √© `plt.bar()`. Leva os seguintes par√¢metros:
- `x`: dados categ√≥ricos para cada barra
- `height`: dados num√©ricos para determinar a altura de cada barra
- `width`: um n√∫mero que podemos passar para definir a largura de cada barra
- `align`: defina como 'center'ou 'edge'para alinhar cada barra no eixo x

````python
bar_data = pd.read_csv('data/plant_data_simplified.csv')
bar_data.head()
````

````python
plt.bar(x = bar_data.PH, height = bar_data.average_leaf_width, width = 0.8, align = 'center')   
plt.title('Effect of pH on Leaf Width')
plt.xlabel('pH')
plt.ylabel('Leaf Width (cm)')
plt.show()
````

#### Adicionar barras de erro

√â comum haver alguma incerteza ou imprecis√£o nas medi√ß√µes, especialmente em medi√ß√µes emp√≠ricas coletadas por observa√ß√£o.
Frequentemente, quantificamos esta incerteza atrav√©s de intervalos de confian√ßa, que levam em conta a dispers√£o e o 
n√∫mero de pontos de dados inclu√≠dos no c√°lculo da nossa estat√≠stica resumida.

Podemos representar visualmente esta incerteza adicionando barras de erro a um gr√°fico. Usaremos a fun√ß√£o geral `
plt.errorbar()` do matplotlib , que recebe par√¢metros para:
- `x` e `y`: reafirmar os valores X e Y do gr√°fico subjacente
- `yerr` e/ou `xerr`: definir valores de erro na dire√ß√£o X ou Y
- `color`: defina a cor da barra de erro (opcional)
- `fmt`: alterar o marcador

````python
plt.bar(x = bar_data.PH, height = bar_data.average_leaf_width, width = 0.8, align = 'center')
plt.title('Effect of pH on Leaf Width')
plt.xlabel('pH')
plt.ylabel('Leaf Width (cm)')
plt.errorbar(bar_data.PH, bar_data.average_leaf_width, yerr = bar_data.error, fmt = 'o', color = 'orangered') 
plt.show()
````

### 2. Gr√°fico de dispers√£o

Os gr√°ficos de dispers√£o s√£o √≥timos para comparar diretamente duas vari√°veis ‚Äã‚Äãnum√©ricas cont√≠nuas, ajudando-nos a 
visualizar a rela√ß√£o ou correla√ß√£o entre elas.

- **Correla√ß√£o positiva**: Se os pontos estiverem espalhados do canto inferior esquerdo para o canto superior direito 
do gr√°fico, poderemos dizer ‚Äú√Ä medida que X aumenta, Y tamb√©m aumenta‚Äù.
- **Correla√ß√£o negativa**: Se os pontos estiverem espalhados do canto superior esquerdo para o canto inferior direito do gr√°fico, poderemos dizer ‚Äú√Ä medida que X aumenta, Y diminui‚Äù.
- **N√£o h√° correla√ß√£o**: Se os pontos estiverem espalhados em uma linha horizontal, podemos dizer ‚Äú√Ä medida que X aumenta, Y permanece constante‚Äù. Uma nuvem de pontos dispersos sem agrupamento direcional claro tamb√©m indica aus√™ncia de correla√ß√£o.

Fun√ß√£o gr√°fica de dispers√£o, `plt.scatter()` com os seguintes par√¢metros:
- `x` e `y`: as vari√°veis ‚Äã‚Äãnum√©ricas cont√≠nuas a serem comparadas
- `color`: cor do marcador, como c√≥digo de cor, nome de cor ou c√≥digo hexadecimal
- `alpha`: opacidade do marcador, como um n√∫mero entre 0 (transparente) e 1 (opaco)

````python
scatter_data = pd.read_csv('data/spotify_data_by_genres.csv')
scatter_data.head()
````

````python
plt.scatter(x=scatter_data.danceability, y=scatter_data.valence)    # comparacao x e y
plt.title('Mood and Danceability correlation in Spotify genres')
plt.xlabel('Danceability')
plt.ylabel('Valence / Mood (sadder to happier)')
plt.show()
````

````python
plt.scatter(x=scatter_data.danceability, y=scatter_data.valence, alpha=0.25)    # add alpha
plt.title('Mood and Danceability correlation in Spotify genres')
plt.xlabel('Danceability')
plt.ylabel('Valence / Mood (sadder to happier)')
plt.show()
````

````python
plt.scatter(x=scatter_data.danceability, y=scatter_data.valence, color='teal', alpha=0.15)
plt.title('Mood and Danceability correlation in Spotify genres')
plt.xlabel('Danceability')
plt.ylabel('Valence / Mood (sadder to happier)')
plt.show()
````

### 5. Histograma

`plt.hist()` para fazer um histograma, um gr√°fico que mostra a dispers√£o de uma vari√°vel de dados cont√≠nuos. Recebe os seguintes par√¢metros:
- `x`: o valor que est√° sendo distribu√≠do (como tamanho do cal√ßado ou altura)
- `bins`: especifica quantas caixas criar (por exemplo, 10) ou onde est√£o as bordas das caixas (como uma lista de valores)
- `range`: a faixa inferior e superior das caixas. Se n√£o for especificado, defina os valores m√≠nimo e m√°ximo para `x`
- `color`: define a cor das barras

````python
%matplotlib inline
plt.rcParams['figure.figsize'] = (6, 4)   
plt.rcParams['figure.dpi'] = 75 

hist_data = pd.read_csv('data/lobster_dist.csv')
hist_data.head()
````

````python
# Criando o histograma e colocando bins
plt.hist(x=hist_data.carapace_length, bins=10)
plt.show()
````

````python
# Adicionando titulo e rotulos
plt.hist(x=hist_data.carapace_length, bins=20)
plt.title('Lobsters tagged by size')    
plt.ylabel('Number tagged')
plt.xlabel('Carapace length (mm)')
plt.show()
````

````python
# Determinadno range e cor
plt.hist(x=hist_data.carapace_length, bins=30, range=(75,155), color='gold')
plt.title('Lobsters tagged by size')
plt.ylabel('Number tagged')
plt.xlabel('Carapace length (mm)')
plt.show()
````

#### Adicionar linhas AB

Uma **linha AB** √© uma linha reta adicionada a um gr√°fico (ou seja, ‚Äúdo ponto A ao ponto B‚Äù). Pode ser vertical, horizontal ou diagonal. Geralmente, uma linha AB ajuda a demarcar uma √°rea de outra ou fornece contexto que ajuda o visualizador a entender os dados no gr√°fico. No matplotlib, a fun√ß√£o plt.axvline()cria uma linha AB vertical e `plt.axhline()` uma linha AB horizontal. Trabalharemos com `plt.axvline()`, que leva os seguintes par√¢metros:

- `x`: onde posicionar a linha ao longo do eixo x
- `ymin`: qu√£o perto da parte inferior do gr√°fico a linha come√ßa. A configura√ß√£o `ymin=0` inicia a linha na parte inferior do gr√°fico, `ymin=.5` inicia-a na metade e `ymin=1` come√ßa na parte superior do gr√°fico. Geralmente definido como 0
- `ymax`: qu√£o perto da parte inferior do gr√°fico a linha termina. A configura√ß√£o `ymax=0` termina a linha na parte inferior do gr√°fico, `ymax=.5` termina na metade e `ymax=1` termina na parte superior do gr√°fico. Geralmente definido como 1
- `linewidth`: largura da linha AB
- `dashes`: padr√£o de tra√ßo dado como (`line_length`, `space_length`)
- `color`: cor da linha AB

Podemos facilmente adicionar anota√ß√µes a um gr√°fico (ou rotular uma linha AB) usando `plt.annotate()`

````python
plt.hist(hist_data.carapace_length, bins=30, range=(75,155), color='gold')
plt.title('Lobsters tagged by size')
plt.ylabel('Number tagged')
plt.xlabel('Carapace length (mm)')
plt.axline((83, 0), (83, 1), linewidth=2, dashes=(1,2), color='mediumblue')
plt.axline((127, 0), (127, 1), linewidth=2, dashes=(1,2), color='mediumblue')
plt.show()
````

````python
plt.hist(hist_data.carapace_length, bins=30, range=(75,155), color='gold')
plt.title('Lobsters tagged by size')
plt.ylabel('Number tagged')
plt.xlabel('Carapace length (mm)')
plt.axline((83, 0), (83, 1), linewidth=2, dashes=(1,2), color='mediumblue')
plt.axline((127, 0), (127, 1), linewidth=2, dashes=(1,2), color='mediumblue')
plt.annotate('Minimum legal catch size', (84, 23), color='mediumblue')
plt.annotate('Maximum legal\ncatch size', (128, 7), color='mediumblue')
plt.show()
````

### 4. Exibir varios gr√°ficos

Quando quisermos fazer compara√ß√µes sem complicar demais um √∫nico gr√°fico, podemos usar faceta√ß√£o , onde fazemos um gr√°fico para cada grupo e os alinhamos um ao lado do outro.

No Seaborn, podemos usar a `sns.FacetGrid()` fun√ß√£o para construir uma grade de gr√°ficos e ent√£o mapear uma fun√ß√£o de plotagem nela. Essas fun√ß√µes recebem um `kind` par√¢metro para especificar o tipo de gr√°fico e, em seguida, adicionam os par√¢metros familiares como `data`, `x`, `y`, `hue` ou `style`.

- `sns.relplot()`plota dados relacionais de m√∫ltiplas vari√°veis, como gr√°ficos '`line`' ou '`scatter`'
- `sns.displot()` plota dados de distribui√ß√£o como histogramas ('`hist`') ou '`kde`' gr√°ficos.
- `sns.catplot()` plota dados categ√≥ricos como '`bar`' ou '`box`' gr√°ficos.

Para criar as facetas, definimos um dos dois par√¢metros para uma vari√°vel de agrupamento: `row` para exibir o gr√°fico verticalmente em linhas ou `col` para exibir os gr√°ficos horizontalmente em colunas.

- `sns.jointplot()` plota duas vari√°veis ‚Äã‚Äãjuntamente com suas distribui√ß√µes individuais nas margens para que possamos verificar rapidamente um relacionamento em v√°rios n√≠veis ao mesmo tempo.

### 5. Use paletas de cores
Seaborn inclui seis paletas de cores qualitativas : profundo, suave, pastel, claro, escuro e dalt√¥nico.



No entanto, se tivermos dados ordenados, talvez seja melhor escolher entre dois tipos de paletas ordenadas:

- **Paletas sequenciais** mudam de cores mais claras para mais escuras em uma dire√ß√£o. Essas paletas podem ser uma boa escolha para categorias ordenadas ou escalas num√©ricas cont√≠nuas.



- **Paletas divergentes** geralmente incluem uma sequ√™ncia sobre duas cores onde os incrementos mudam de escuro para claro no centro para escuro novamente no final. Essas paletas podem ser uma boa escolha para mostrar valores absolutos para dados que incluem valores negativos e positivos, ou dados onde duas coisas s√£o opostas, como partidos pol√≠ticos ou resultados favor√°veis ‚Äã‚Äãe desfavor√°veis.



#### **Paletas personalizadas**
Criar uma paleta personalizada com `sns.color_palette()`. Por exemplo, para fazer uma paleta de cores sequencial de azuis do claro ao escuro em incrementos de 10, podemos usar o seguinte c√≥digo:
```python
    seqBlues = sns.color_palette("light:blue", n_colors=10)
```

Podemos transformar nossa `seqBlues` paleta em um mapa de cores cont√≠nuo matplotlib trocando `n_colors=10` por `as_cmap=True`. Outras fun√ß√µes de paleta com objetivos mais espec√≠ficos tamb√©m podem ser usadas para cria√ß√£o de paleta, como `sns.light_palette()` ou s`ns.diverging_palette()`.

#### **Aplicando paletas**
Podemos alterar a paleta de um gr√°fico definindo o `palette` par√¢metro com o nome da paleta. Os nomes das paletas pr√©-fabricadas s√£o fornecidos entre aspas (`palette='dark'`), enquanto as paletas personalizadas s√£o fornecidas sem aspas (`palette=seqBlues`).

Aplicar configura√ß√µes a todos os gr√°ficos executando uma fun√ß√£o no in√≠cio do nosso notebook:
- `sns.set_palette()` aplica a mesma paleta de cores a todos os gr√°ficos.
- `sns.set_theme()` define os mesmos par√¢metros de plotagem para todos os gr√°ficos, incluindo recursos como bordas, grades e cor de fundo.

### 6. Adicione r√≥tulos e anota√ß√µes
Cores e estilos nos ajudam a criar gr√°ficos claros e informativos. Podemos adicionar ainda mais clareza e informa√ß√µes especializadas com t√≠tulos, r√≥tulos e anota√ß√µes. √â essencial apra ajudar nosso p√∫blico a aproveitar ao m√°ximo os dados.

#### **Parcela √∫nica**
Para gr√°ficos em n√≠vel de eixo, como `sns.lineplot()` ou `sns.boxplot()`, podemos adicionar rapidamente um t√≠tulo e r√≥tulos de eixo como par√¢metros de `.set()`. Tamb√©m podemos adicionar anota√ß√µes de texto em um ponto espec√≠fico `(x, y)` usando `.annotate()` o n√≠vel dos eixos.

#### **V√°rios lotes**
Para gr√°ficos de n√≠vel de figura, como `sns.relplot()` ou `sns.catplot()`, definimos os r√≥tulos de eixo para todos os gr√°ficos na figura usando `.set_axis_labels()`.

Podemos usar `.set_titles()` um `{col_name} `modelo para fornecer programaticamente um t√≠tulo em n√≠vel de eixo para cada gr√°fico da figura. Para t√≠tulo principal, podemos usar `.fig.subtitle()`

````python
waste = pd.read_csv('data/waste.csv')
waste['income'] = waste['income'].astype('category').cat.set_categories(['Low', 'Lower Middle', 'Upper Middle', 'High'], ordered=True)
%matplotlib inline
plt.rcParams['figure.dpi'] = 72
waste.head()
````

````python
# plot
p = sns.barplot(data=waste, x='msw', y='income', errorbar=None, palette='Blues')
# labels
p.set(title="Average Per Capita MSW Increases with Country Income Level", 
      xlabel="Average Municipal Solid Waste (MSW) \n(tons per year per person)", 
      ylabel="")
plt.show()
````

````python
mean_msw = waste.groupby('income')['msw'].mean().round(2)
print(mean_msw)
````

````python
p = sns.barplot(data=waste, x='msw', y='income', errorbar=None, palette='Blues')

# labels
p.set(title='Average Per Capita MSW Increases with Country Income Level', 
      xlabel='Average Municipal Solid Waste (MSW) \n(tons per year per person)', 
      ylabel='')

# annotations
p.annotate(text="0.16", xy=(0.17,0), weight='bold')
p.annotate(text="0.22", xy=(0.23,1), weight='bold')
p.annotate(text="0.30", xy=(0.31,2), weight='bold')
p.annotate(text="0.61", xy=(0.62,3), weight='bold')

plt.show()
````

````python
# plot
p = sns.barplot(data=waste, x='msw', y='income', ci=None, palette='Blues')

# labels
p.set(title='Average Per Capita MSW Increases with Country Income Level', 
      xlabel='Average Municipal Solid Waste (MSW) \n(tons per year per person)', 
      ylabel='')

# annotations
for i,msw in enumerate(mean_msw):
    p.annotate(text=str(msw), xy=(msw+0.01,i), weight='bold')
    
p.spines['top'].set_visible(False)
p.spines['right'].set_visible(False)

plt.show()
````
